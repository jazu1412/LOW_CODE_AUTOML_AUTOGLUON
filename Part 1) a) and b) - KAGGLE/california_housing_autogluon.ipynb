## Data Loading

Now, let's load the California Housing dataset. We'll use Kaggle API to download the dataset.

```python
from sklearn.model_selection import train_test_split
import pandas as pd

directory = 'california-house-prices/'
label = 'Sold Price'

eval_metric = 'rmse'


# Split into train and test sets
train_data= pd.read_csv(directory+'train.csv')
test_data = pd.read_csv(directory+'test.csv')
```

```python
# Specify the label column (the target variable)
label = 'Sold Price'

# Create the predictor
predictor = TabularPredictor(label=label,eval_metric='rmse', problem_type='regression').fit(
    train_data=train_data,
    presets='best_quality',  # Could also be 'medium_quality_faster_train', 'good_quality_faster_inference_only_refit', etc.
    time_limit=600  # Limit to 600 seconds (10 minutes)

)
```

## Model Training

We'll use AutoGluon's TabularPredictor to train our model. We'll use the 'multimodal' preset for hyperparameters and set a time limit for training.
